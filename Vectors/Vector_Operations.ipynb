```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
plt.quiver(0,0,2,9)
plt.show()
```

```python
plt.quiver(0,0,2,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-4,-6,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Addition of two vectors

```python
vector_1=np.asarray([0,0,2,5])
vector_2=np.asarray([0,0,-3,3])
sum=vector_1+vector_2
print(sum)
```

```python
plt.quiver(0,0,2,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,6,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,-1,8,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Subtraction of vectors

```python
vector_1=np.asarray([0,0,2,5])
vector_2=np.asarray([0,0,-3,3])
sum=vector_1-vector_2
print(sum)
```

```python
plt.quiver(0,0,2,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,6,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(0,0,5,2,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Multiplying the vector with scalar

```python
vector_1=np.asarray([0,0,2,5])
vector_2=2*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,4,10,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
vector_1=np.asarray([0,0,2,5])
vector_2=-0.5*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,2,5,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-1,-2.5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
